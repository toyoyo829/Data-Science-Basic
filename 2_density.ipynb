### ガウス分布の混合を関数化してみましょう

```python
x = np.linspace(-5,5,200) #プロットするx座標を定義
```

```python
# n:混合数 ms:混合する各分布の平均のリスト h:各分布の分散(共通)
def gaussian_mixture(n,ms,h):
    z = 0 * x
    for m in ms:  
        plt.plot(x, normpdf(x,m,h)/n, color="blue")
        「zにnormpdf(x,m,h)/nを足す処理を書いてください。」
    return z
```

```python
#動作確認をしてみましょう。
```

```python
#分散0.01の場合
plt.plot(x, gaussian_mixture(5,[-2,0,1,1.5,2],0.01), color="red")
plt.show()
```

```python
#分散0.5の場合
plt.plot(x, gaussian_mixture(5,[-2,0,1,1.5,2],0.5), color="red")
plt.show()
```

```python
#分散10の場合
plt.plot(x, gaussian_mixture(5,[-2,0,1,1.5,2],10), color="red")
plt.show()
```
